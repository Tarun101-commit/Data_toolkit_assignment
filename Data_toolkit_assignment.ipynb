```python
# 4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap. 

from matplotlib.colors import Colormap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sympy import true

data = {
    'Employee ID': ['EMP001', 'EMP002', 'EMP003', 'EMP004', 'EMP005', 'EMP006', 'EMP007'],
    'Employee Name': ['Alice Smith', 'Bob Jones', 'Charlie Brown', 'Diana Prince', 'Evan Wright', 'Fiona Gallagher', 'George Clark'],
    'Age': [25, 34, 41, 29, 52, 22, 38],
    'Experience': [2.5, 9.0, 15.0, 4.5, 22.0, 1.0, 11.5],
    'Salary': [54000, 85500, 110000, 68000, 145000, 48000, 92000]
}


df=pd.DataFrame(data)

# correlation
corr_matrix=df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.4f', linewidths=0.8)

plt.title('Corelation Matrix Heatmap')
plt.show()
```

```python
#  Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt 

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

sales_2024 = [100, 120, 140, 160, 180]
sales_2025 = [110, 130, 150, 170, 200]

plt.plot(months,sales_2024, marker='o', linestyle='-', label='sales_2024')
plt.plot(months,sales_2025, marker='s', linestyle='--', label='sales_2025')



plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales by Months')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# 10.   Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt

salary = [30000, 35000, 40000, 45000,
          50000, 55000, 60000, 65000,
          70000, 75000]

sns.histplot(salary, kde=True, bins=6)

plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Freequency')

plt.show()
```

```python
# 11.Perform matrix multiplication using NumPy.

#@ and np.dot() this tww are usually used in matroix multiplication 


import numpy as np

## Rule for matrix  multiplication- Number of Columnn in the first matrix should be equal to the number of rows in the second  matrix.

A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

result=np.dot(A,B)

print(result)
```
